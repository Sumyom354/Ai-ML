<h4>Histogram</h4>

```python
#Create a histogram of 'Age' distribution
plt.figure(figsize=(5,5))
plt.hist(diabetes_model['Age'],color='green',edgecolor='red')
plt.title('Age distribution',fontsize=12)
plt.xlabel('Age of Patients')
plt.ylabel('Freequency of Patients')
```

<h4>Heatmap</h4>

```python
#Create a heatmap
# Calculate the correlation matrix
correlation_matrix = diabetes_model.corr()
# Plot the heatmap
plt.figure(figsize=(8,8))
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

<h4>Boxplot</h4>

```python
#convert numerical 'Outcome' to categorical 'Outcome'
diabetes_model['Outcome']=diabetes_model['Outcome'].astype('category')
```

```python
diabetes_model.info()
```

```python
# Boxplot for 'Glucose' vs 'Outcome'
plt.figure(figsize=(5,5))
sns.boxplot(x='Outcome', y='Glucose', data=diabetes_model,palette='coolwarm')
plt.title('Glucose Levels vs Outcome')
plt.xlabel('Diabetes Outcome (0: No Diabetes, 1: Diabetes)')
plt.ylabel('Glucose Level')
plt.show()
```
